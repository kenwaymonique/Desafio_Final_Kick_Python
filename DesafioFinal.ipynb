```python
df = pd.read_csv('plano_sp_leitos_internacoes_serie_nova_variacao_semanal.csv',encoding='utf-8', delimiter=';')
```

```python
new = df.tail(18)
new
```

```python
cidade = new.nome_drs.value_counts(dropna=True)
cid = cidade.to_frame()
cidd = cidade.index
cidd
```

```python
paciente = new.total_covid_enf_ultimo_dia.value_counts(dropna=True)
pacientess = paciente.to_frame()
ultimopaciente = pacientess.index
ultimopaciente
```

```python
plt.figure(figsize=(9,5))
plt.barh(cidd, ultimopaciente)
plt.xlabel("Indivíduos")
plt.ylabel("Municípios de São Paulo")
plt.title("Indivíduos com Covid na enfermaria nos Municípios de São Paulo em 18-11-2022")
plt.show()
plt.savefig('individuos.png', format='png')
```
